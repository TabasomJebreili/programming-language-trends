# Programming Language Trends on StackOverflow

This project analyzes the monthly post count of different programming languages on StackOverflow from 2008 onward.  
By reshaping and visualizing the data, I explore how the popularity of languages like Python, Java, C++, and Swift has changed over time.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Exploration

```python
df = pd.read_csv('QueryResults.csv', names=['DATE', 'TAG', 'POSTS'], header=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df.count()
```

Total number of post per language.


```python
sum = df.groupby('TAG').sum()
sum.sort_values('POSTS')
```

Number of active months per programming language to identify which had the fewest recorded entries.

```python
dates = df.groupby('TAG').count()
dates.sort_values('DATE')
```

## Data format


```python
pd.to_datetime(df['DATE'][1])
```

```python
type(pd.to_datetime(df['DATE'][1]))
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.head()
```

## Data Manipulation



```python
reshaped_df = df.pivot(index='DATE', columns='TAG', values='POSTS')
```

```python
reshaped_df.shape
```

```python
reshaped_df.tail()
```

```python
reshaped_df.columns
```

```python
reshaped_df.count()
```

Number of entries per programming language:

```python
reshaped_df.fillna(0, inplace=True)
```

```python
reshaped_df.head()
```

```python
reshaped_df.count()
```

```python
reshaped_df.isna().values.any()
```

## Data Visualisaton


```python

plt.plot(reshaped_df.index ,"java", data=reshaped_df)
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index ,"java", data=reshaped_df)
```

two line:

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0,35000)
for column in reshaped_df.columns:
  plt.plot(reshaped_df.index ,reshaped_df[column], linewidth=3, label=reshaped_df[column].name)
plt.legend(fontsize=16)
```

# Smoothing out Time Series Data


```python
roll_df = reshaped_df.rolling(window=12).mean()
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0,35000)
for column in roll_df.columns:
  plt.plot(roll_df.index, roll_df[column], linewidth=3 , label=roll_df[column].name)
plt.legend()
```
